```python
import geopandas as gpd
import pandas as pd

def old_load_kp(data_root):
    df = pd.read_csv(data_root / "CAMELS_FR_attributes/static_attributes/CAMELS_FR_site_general_attributes.csv", sep=";")
    gdf = gpd.read_file(data_root / "./CAMELS_FR_geography/CAMELS_FR_gauge_outlet.gpkg")
    gdf["area"]=df["sit_area_topo"]
    gdf = gdf.to_crs("EPSG:4326")
    return gdf

def load_kp(data_root):
    att = pd.read_csv(data_root / "CAMELS_FR_attributes/static_attributes/CAMELS_FR_station_general_attributes.csv", sep=";")
    df = att[['sta_x_w84', 'sta_y_w84', 'sta_x_w84_snap', 'sta_y_w84_snap', 'sta_area_snap', "sta_code_h3"]]
    gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['sta_x_w84_snap'], df['sta_y_w84_snap']),
                           #index=att["],
                           crs="EPSG:4326")
    return gdf.rename({"sta_area_snap":"area", 
            "sta_x_w84_snap":"x_snap",
            "sta_y_w84_snap":"y_snap",
            "sta_x_w84":"x",
            "sta_y_w84":"y",
           }, axis=1).to_crs("EPSG:4326")
```

```python
def apply_all_modifications(gauge_details, all_modif_pathes):
    gauge_details = gauge_details.copy()
    for modif in all_modif_pathes:
        gauge_details = app.apply_modifications(gauge_details, modif)
    gauge_details["geometry"] = gpd.points_from_xy(gauge_details.lon, gauge_details.lat)
    return gauge_details
```
